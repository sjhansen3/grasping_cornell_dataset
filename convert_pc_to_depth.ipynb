```python
import matplotlib.pyplot as plt
%matplotlib inline
from pyntcloud import PyntCloud

import numpy as np
import pandas as pd
from perception import DepthImage, CameraIntrinsics
```

```python
## Set the camera intrinsics
# Kinect parameters from NYU dataset, used to be close to my kinect data

# Depth Intrinsic Parameters
fx_d = 5.8262448167737955e+02;
fy_d = 5.8269103270988637e+02;
cx_d = 3.1304475870804731e+02;
cy_d = 2.3844389626620386e+02;

# RGB Distortion Parameters
k1_d = -9.9897236553084481e-02;
k2_d = 3.9065324602765344e-01;
p1_d = 1.9290592870229277e-03;
p2_d = -1.9422022475975055e-03;
k3_d = -5.1031725053400578e-01;

# % 3D Translation
t_x = -1500;
t_z = -800;
t_y = -500;

intrinsics = CameraIntrinsics("kinect", fx=fx_d, fy=fy_d, cx=cx_d, cy=cy_d, skew=0.0, height=480, width=620)
```

```python
def get_depth_image(cloud):
    #camera image height and width
    W = intrinsics.height;
    H = intrinsics.width;

    X_world = cloud.points.values[:,0]+t_x
    Y_world = cloud.points.values[:,1]+t_y
    Z_world = -cloud.points.values[:,2]+t_z
    depth = np.sqrt((X_world**2+Y_world**2+Z_world**2))
    #color = 

    X_plane = (X_world * intrinsics.fx / Z_world)
    Y_plane = (Y_world * intrinsics.fy / Z_world)
    
    # Finally, project back onto the RGB plane.
    X_plane = np.round(X_plane).flatten().astype('int')
    Y_plane = np.round(Y_plane).flatten().astype('int')

    #find the indicies that will hit the image plane
    goodInds = np.where(np.logical_and(np.logical_and(np.logical_and(X_plane > 0, X_plane < W), Y_plane > 0), Y_plane < H))
    goodInds = goodInds[0]
    
    #sort the remaining depth values to be projected back
    to_sort = -depth[goodInds]
    depthSorted = np.sort(to_sort);
    order = np.argsort(to_sort)
    depthSorted = -depthSorted;
    
    depthOut = np.zeros((H,W))
    # Z-buffer projection
    for ii in range(len(order)):
        depthOut[Y_plane[goodInds[order[ii]]], X_plane[goodInds[order[ii]]]] = depthSorted[ii]

    # Fix weird values...
    # depthOut(depthOut > maxDepth) = maxDepth;
    depthOut[depthOut < 0] = 0;
    # depthOut[np.isnan[depthOut]] = 0;
#     depthOut = np.rot90(depthOut);
    return depthOut
```

```python
depth_out = get_depth_image(cloud)
plt.imshow(depth_out,cmap='gray')
```
